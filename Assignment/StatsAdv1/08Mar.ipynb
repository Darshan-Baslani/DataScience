# Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point..

```python
import math
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
def pdf(x, mean, sd):
    return norm.pdf(x, loc=mean, scale=sd)


mean = 0
sd = 1
x = 1.5
print(f"PDF of {x} with mean: {mean} and standard deviation: {sd} is {pdf(x,mean,sd)}")

# Graph showing pdf for values -5 to 5
x_values = np.linspace(-5, 5)
pdf_values = [pdf(x, mean, sd) for x in x_values]
plt.plot(x_values, pdf_values)
plt.grid()
plt.show()
```

# Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied..

The properties of a Binomial distribution are as follows:

Discrete Distribution: The Binomial distribution is a discrete probability distribution, meaning it deals with random variables that can only take on whole number values.

Fixed Number of Trials: The distribution models the number of successes in a fixed number of independent Bernoulli trials (experiments), where each trial can result in either success or failure.

Independent Trials: Each trial's outcome does not depend on the outcomes of the previous or subsequent trials. The trials are considered to be independent.

Constant Probability: The probability of success (denoted by 'p') remains constant for each trial, and the probability of failure is given by (1 - p).

Binomial Coefficient: The probability of getting 'k' successes in 'n' trials is given by the binomial coefficient, denoted as C(n, k) or "n choose k," which is equal to n! / (k! * (n - k)!), where 'n!' denotes the factorial of 'n'.

Two examples where the Binomial distribution can be applied are:

Example 1 - Coin Toss: Suppose you flip a fair coin (with an equal probability of getting heads or tails) 10 times. The Binomial distribution can be used to calculate the probability of getting a specific number of heads (successes) out of those 10 flips.

Example 2 - Quality Control: A factory produces light bulbs, and each light bulb has a 95% chance of passing quality control (success) and a 5% chance of being defective (failure). If the factory produces 100 light bulbs, the Binomial distribution can be applied to calculate the probability of a specific number of defective bulbs in the batch.

# Q7. How Binomial distribution different from Poisson distribution?

The binomial distribution is one, whose possible number of outcomes are two, i.e. success or failure. On the other hand, there is no limit of possible outcomes in Poisson distribution

# Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variancee

```python
from scipy.stats import poisson

mean = 5
poisson = np.array(np.random.poisson(mean, 1000))
sample_mean = np.mean(poisson)
variance = np.var(poisson)

print(f"Sample Mean : {sample_mean}")
print(f"Variance : {variance}")
```
